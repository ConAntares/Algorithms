# Linear Dynamical Systems

## Basic Linear Dynamical System
Let's simulate a time-invariant linear dynamic system  
$x_{t+1}=Ax_t, t=1,...,T$  
with dynamics matrix  
$A= \begin{bmatrix}
    0.97 &  0.10 & -0.05 \\
   -0.30 &  0.99 &  0.05 \\
    0.01 & -0.04 &  0.96
    \end{bmatrix}$  
and initial state $x_1=(1,0,-1)$. We store the state trajectory in the $n \times T$ matrix `state_traj`, with the $i$th column $x_t$. 

```julia
using Plots; pyplot()

x1 = [ 1, 0, -1] # Initinal state
n = length(x1)
T = 50
A = [ 0.97  0.10 -0.05;
     -0.30  0.99  0.05;
      0.01 -0.04  0.96]
state_traj = [x1 zeros(n,T-1)]

# Dynamics recursion
for t=1:T-1
    state_traj[:,t+1] = A*state_traj[:,t]
end

plot(1:T, state_traj'[:,1], xlabel="t", label="xt_1")
plot!(1:T, state_traj'[:,2], xlabel="t", label="xt_2")
plot!(1:T, state_traj'[:,3], xlabel="t", label="xt_3")
```
