# Julia Plots

## 2D Plotting with PyPlot backend

```julia
using Plots; pyplot()
using LaTeXStrings

PyPlot.rc("text", usetex = "true")
PyPlot.rc("font", family = "CMU Serif")

X = range(-2pi, stop=2pi, length=100)
Y = 1.5 .* sin.(0.3 .+ 0.7X)

plot(dpi=256, gridls=:dash, legend=:topright, frame=:both,
     fgborder="#505050", fgaxis="#505050", tickfontsize=10, legendfontsize=10)
plot!(X,Y, title = L"Title $\alpha$", lw=2, label=L"wave $f=f(x)$", annotation=(1, 1, L"Annotation $\beta$"))
```
