# Plots

## 2D Plotting

```julia
using Plots; pyplot(); 
using LaTeXStrings
```

```julia
X = range(-2pi, stop=2pi, length=100)
Y = 1.5 .* sin.(0.3 .+ 0.7X)

plot(fontfamily="CMU Serif", dpi=512, gridls=:dash, legend=:topright, frame=:both)
plot!(X,Y, title = raw"ABC $\alpha$", lw=2, label=raw"wave $f=f(x)$", annotation=(1, 1, L"\beta",))
```

## 3D Plotting

```julia
n = 100
ts = range(0, stop=8π, length=n)
x = ts .* map(cos, ts)
y = (0.1ts) .* map(sin, ts)
z = 1:n
plot(x, y, z, zcolor=reverse(z), m=(10, 0.8, :blues, Plots.stroke(0)), leg=false, cbar=true, w=5)
plot!(zeros(n), zeros(n), 1:n, w=10)
```
