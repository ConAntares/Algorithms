# Plots

## PyPlot

```julia
using Plots; pyplot(); 
using LaTeXStrings
```

```julia
X = range(-2pi, stop=2pi, length=100)
Y = 1.5 .* sin.(0.3 .+ 0.7X)

plot(fontfamily="CMU Serif", dpi=512, gridls=:dash, legend=:topright, frame=:axes)
plot!(X,Y, title = raw"ABC $\alpha$", lw=2, label=raw"wave $f=f(x)$", annotation=(1, 1, L"\beta",))
```
